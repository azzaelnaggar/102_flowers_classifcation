# import libraries

```python
import os
import numpy as np 
import shutil
from scipy.io import loadmat
import splitfolders
```

# sort image by its name

```python
file_list = os.listdir("102flowers\jpg")
#grab last 4 characters of the file name: and sort files 
def last_4chars(x):
    return(x[-4:])
images_list=sorted(file_list, key = last_4chars)   
images_list
```

# load mat file

```python
load_ = loadmat(r"imagelabels.mat")
lables=["labels"][0]
```

```python
labels_list=labels["labels"].tolist()
```

# get the labels and create folder for each one

```python
folders_names = np.unique(labels_list)
```

```python
# make folder for each class
for folder in folders_names:
    path = os.path.join("data/", str(folder)) 
    os.makedirs(path)
```

# copy image to its corresponding folder(class)

```python
for image ,fol in zip(images_list,labels_list):
    shutil.copy('102flowers/jpg/'+image,os.path.join('data',str(fol)))
```

# split the data into train,validation,test data

```python
splitfolders.ratio("data",seed=1337, output="data-Splitted", ratio=(0.8, 0.1, 0.1))
```
